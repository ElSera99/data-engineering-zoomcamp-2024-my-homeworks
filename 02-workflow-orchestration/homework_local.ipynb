# Import libraries

Libraries needed to process data

````python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
````

# Load Dataframes

A list of the compressed CSV files in order to load into a single dataframe

````python
URLs = [
    "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2020-10.csv.gz",
    "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2020-11.csv.gz",
    "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_2020-12.csv.gz"
]
````

An empty data frame for initialization

````python
df = pd.DataFrame()
````

Append each data frame created from each URL into a single data frame

````python
for element in URLs:
    df_tmp = pd.read_csv(element, compression="gzip", low_memory=False)
    df = pd.concat([df,df_tmp])    
````

````python
df.head(10)
````

````python
df.tail(10)
````

## Transform columns

Transform columns into desired format such as:
- Set column names as lower case
- Replace all spaces from column names to underscore
- Set correct ID names as snake case

````python
df_original_columns = df.columns
df.columns
````

````python
df.columns = (df.columns
              .str.lower()
              .str.replace(" ", "_")
              .str.replace("id","_id")
              .str.replace("pul","pu_l")
              .str.replace("dol","do_l")
)
````

````python
df.columns
````

## Set date column
<a id='set_date_colum'></a>

Parse datetime columns from string into datetime

````python
df["lpep_pickup_datetime"] = pd.to_datetime(df["lpep_pickup_datetime"])
df["lpep_dropoff_datetime"] = pd.to_datetime(df["lpep_dropoff_datetime"])
````

````python
df["lpep_pickup_datetime"]
````

````python
df["lpep_dropoff_datetime"]
````

## Create date column

````python
df["lpep_pickup_date"] = df["lpep_pickup_datetime"].dt.date
````

````python
df["lpep_pickup_date"]
````

## Filter data

Filter rows that do not have passengers nor trip distance

````python
df_filtered = df[(df["passenger_count"] > 0) & (df["trip_distance"] > 0)]
df_filtered
````
