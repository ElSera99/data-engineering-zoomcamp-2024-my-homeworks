# Import Libraries

```python
import pandas as pd
import numpy as np
```

# Read Data

```python
filename = 'fhv_tripdata_2019-01.csv.gz'
```

```python
df = pd.read_csv(filename, compression='gzip', low_memory=False)
```

```python
df
```

```python
df.columns
```

# Data and Types Analysis

```python
df.dtypes
```

## Is there null values?

```python
df.isnull().any()
```

## Unique values

```python
df['dispatching_base_num'].unique()
df['dispatching_base_num'].replace('\\N', np.nan,inplace=True)
```

```python
sorted(df['PUlocationID'].unique())
```

```python
sorted(df['DOlocationID'].unique())
```

```python
sorted(df['SR_Flag'].unique())
```

```python
list(df['Affiliated_base_number'].unique())
df['Affiliated_base_number'].replace('\s+', '',regex=True,inplace=True)
list(df['Affiliated_base_number'].unique())
```

## Replace Null values

```python
df[['PUlocationID', 
'DOlocationID', 
'SR_Flag', 
'Affiliated_base_number']].fillna(np.nan, inplace=True)
```

## Correct Data Types

```python
df['dispatching_base_num'] = df['dispatching_base_num'].astype('string')
df['PUlocationID'] = df['PUlocationID'].astype('Int64')
df['DOlocationID'] = df['DOlocationID'].astype('Int64')
df['SR_Flag'] = df['SR_Flag'].astype('Int64')
df['Affiliated_base_number'] = df['Affiliated_base_number'].astype('string')

# df.astype({
#             'dispatching_base_num': str,
#             'PUlocationID': int,
#             'DOlocationID': int, 
#             'SR_Flag': int, 
#             'Affiliated_base_number': str
#             }, errors='ignore') 
```

## Datetime transformation

```python
df['pickup_datetime'] = pd.to_datetime(df['pickup_datetime'], format="%Y-%m-%d %H:%M:%S")
```

```python
df['dropOff_datetime'] = pd.to_datetime(df['dropOff_datetime'], format="%Y-%m-%d %H:%M:%S")
```
